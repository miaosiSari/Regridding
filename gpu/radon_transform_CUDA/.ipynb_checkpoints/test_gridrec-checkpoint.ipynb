```python
import tomopy
obj = tomopy.shepp2d(1024) # Generate an object.
ang = tomopy.angles(180) # Generate uniformly spaced tilt angles.
sim = tomopy.project(obj, ang) # Calculate projections.
# Reconstruct object:
rec = tomopy.recon(sim, ang, algorithm='fbp',filter_name='ram-lak')
import pylab
pylab.imshow(obj[0], cmap='gray')
pylab.show()
pylab.imshow(rec[0], cmap='gray')
pylab.show()
```
